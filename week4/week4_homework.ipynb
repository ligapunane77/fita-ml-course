## **Kursa pārskats: W1–W3 refleksija**

### *Kas bija visvieglāk?*
Klasifikācija (W1) bija visintuitīvākā — binārā atbilde (pirks/nepirks) ir viegli saprotama, un F1 score deva skaidru priekšstatu par modeļa kvalitāti. Random Forest darbojās labi gandrīz bez parametru pielāgošanas.

### *Kas bija visgrūtāk?*
Regresija (W2) sagādāja lielākās grūtības — PageValues sadalījums ar daudzām nullēm padarīja lineārās regresijas uzdevumu sarežģītu. R² = 0.046 parādīja, ka lineārs modelis nespēj uztvert šo datu struktūru.

### *Ko darītu citādi?*
Ja sāktu no sākuma, vairāk laika veltītu EDA posmam pirms modeļa izvēles — īpaši target mainīgā sadalījuma izpētei. Regresijā agrāk saprastu, ka PageValues nav labs target lineārai regresijai un izvēlētos log-transformāciju.

### *Kura metrika bija visintuitīvākā?*
F1 score — tā apvieno precision un recall vienā skaitlī un ir saprotama biznesa kontekstā: cik no īstajiem pircējiem modelis atrada. R² bija grūtāk interpretējams — negatīvs vai tuvu nullei R² nebūt nenozīmē, ka modelis kļūdās absolūtos skaitļos.

### *Data leakage un overfitting*
Data leakage tika novērsts ar Pipeline — StandardScaler tika fit tikai uz treniņa datiem. Overfitting nebija būtiska problēma Random Forest modelim (CV std = 0.043 — stabils rezultāts).

## *Kāpēc Pipeline un Cross-Validation ir svarīgi?*
**Pipeline** novērš data leakage — scaler tiek fit tikai uz treniņa datiem katrā fold, nevis uz visiem datiem uzreiz.
**Cross-Validation** dod uzticamu modeļa novērtējumu — vidējais no 5 neatkarīgiem eksperimentiem, nevis viens nejaušs split. Bez tiem rezultāti var būt optimistiski neprecīzi.

## **Ieteiktais modelis**

Random Forest klasifikators ar Pipeline - stabils rezultāts ar mazu standarta novirzi, kas nozīmē uzticamu darbību arī uz jauniem datiem. Klasterizācija ir noderīga segmentācijai, bet nevar atbildēt uz tiešo biznesa jautājumu — "Vai klients nopirks preci?" — tāpēc Random Forest ir visnoderīgākais biznesam.
